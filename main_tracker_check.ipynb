```python
#
import os
import random

import cv2
from ultralytics import YOLO
from deepsort_tracker import Tracker


# video_path = os.path.join('videos', 'city.mp4')
video_path = r'C:\Users\bunny\Downloads\VIS_Onshore\Videos\MVI_1469_VIS_encoded.mp4'
video_out_path = r'C:\Users\bunny\Downloads\VIS_Onshore\Videos\12MVI_1584_VIS_encoded.mp4'

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()

cap_out = cv2.VideoWriter(video_out_path, cv2.VideoWriter_fourcc(*'MP4V'), cap.get(cv2.CAP_PROP_FPS),
                          (frame.shape[1], frame.shape[0]))

model = YOLO(r"C:\Users\bunny\Desktop\SORT-DeepSORT-Tracker-main\runs\detect\100\best.pt")
model.fuse()

tracker = Tracker()

colors = [(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) for j in range(10)]

detection_threshold = 0.3
while ret:

    results = model(frame, verbose=False)

    for result in results:
        detections = []
        for r in result.boxes.data.tolist():
                x1, y1, x2, y2, score, class_id = r
                x1 = int(x1)
                x2 = int(x2)
                y1 = int(y1)
                y2 = int(y2)
                class_id = int(class_id)
                if score > detection_threshold:
                    detections.append([x1, y1, x2, y2, score])

        tracker.update(frame, detections)

        for track in tracker.tracks:
            bbox = track.bbox
            x1, y1, x2, y2 = bbox
            track_id = track.track_id
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (colors[track_id % len(colors)]), 3)
            label = f"ID: {track_id}"
            
            # Фон для текста
            (text_width, text_height), baseline = cv2.getTextSize(
                label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 2)
            
            # Рисуем фон для текста
            cv2.rectangle(frame, 
                         (int(x1), int(y1) - text_height - 10),
                         (int(x1) + text_width, int(y1)),
                         (colors[track_id % len(colors)]),
                         -1)
            
            # Выводим текст
            cv2.putText(frame, label,
                       (int(x1), int(y1) - 5),
                       cv2.FONT_HERSHEY_SIMPLEX, 0.6,
                       (255, 255, 255), 2)
            


    cap_out.write(frame)
    ret, frame = cap.read()

cap.release()
cap_out.release()
cv2.destroyAllWindows()
```
